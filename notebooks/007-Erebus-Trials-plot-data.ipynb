```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drive='e:'
campaign='SDA-trials-2023'


# AUV log files read and get DF
folder = 'files'
subfolder='log'
subfolder='processed-log-data'
auvfile=os.path.join(drive,campaign,folder,subfolder,'auvdata.csv')
auv=pd.read_csv(auvfile)

# USBL log files read and get DF
folder='USBL'
subfolder='data'
usblfile=os.path.join(drive,campaign,folder,subfolder,'USBL-ObsLatLong.csv')
usbl=pd.read_csv(usblfile)
```

```python
auv["time-nav"]=pd.to_datetime(auv["time-nav"])
```

```python
mask = (auv['time-nav'] > '2023 08 01') & (auv['time-nav'] <= '2023 08 04')
gauv=auv.loc[mask]
```

```python
## plot USBL vs. AUV lat Lon

x=usbl.Lon.where(usbl.Address=='2207')
y=usbl.Lat.where(usbl.Address=='2207')



X=gauv.Lon
Y=gauv.Lat

plt.scatter(X,Y,
            c='blue', 
            marker='.',
            s=10,
            label='AUV DR',
            alpha=0.3, edgecolors='none')
plt.scatter(x,y,
            c='deeppink', 
            marker='.',
            s=50,
            label='Nano 2207',
            alpha=0.9, edgecolors='none')

plt.title('Erebus Trials, Gairloch AUV missions')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)

plt.savefig(os.path.join(drive,campaign,'figures','AUVvsUSBL.png'))
```

```python
## plot 
auv.plot(kind='scatter',
        x='Lon',
        y='Lat',
        s=5,
        c='altitude',
        colormap='plasma')
plt.title('Erebus Trials, Gairloch AUV missions')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.savefig(os.path.join(drive,campaign,'figures','AUValtitude.png'))
```

```python
X=auv.salinity
Y=auv.temperature_x
z=auv.pressure
auv.plot(kind='scatter',
            x='salinity',
            y='temperature_x',
            s=5,
            c='pressure',
            colormap='viridis')
plt.xlim([33, 35])
plt.title('Erebus Trials, AUV missions')
plt.xlabel('Salinity (PSU)')
plt.ylabel('Temperature (degC)')
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(drive,campaign,'figures','AUVTS.png'))
```
