```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
indir = os.path.join('e:','ECOWind-accelerate','image-metadata')
```

```python
project='ECOWind-accelerate'
drive ='e:'
filename = os.path.join(drive,project,'image-metadata','photo-coords.csv')
df=pd.read_csv(filename)
fig,ax = plt.subplots(figsize=[4,4])
alt = df["altitude"]
pit = df["pitch"]
ax.plot(pit,alt,'.',color='dimgray')
ax.plot([-20,20],[1.8,1.8],label='desired altitude')
ax.set_xlim(-20, 20)
ax.set_ylim(1, 4)
ax.set_xlabel('Pitch')
ax.set_ylabel('Altitude')
plt.legend()
plt.grid()
plt.savefig(os.path.join(drive,project,'alt-pitch.png'),dpi=150)
```
