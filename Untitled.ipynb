```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cmocean.cm as cmo
```

```python
## PLOT CTD DATA

project='ECOWind-accelerate'
drive ='e:'
filename = os.path.join(drive,project,'master','files','processed-log-data','CTDdata.csv')
df=pd.read_csv(filename,index_col='time')
df
```

```python
fig,ax = plt.subplots(figsize=[8,4])
pres = df["pressure"]
sal = df["salinity"]
ax.plot(pres,sal,'.',color='dimgray')
ax.set_xlim(5, 35)
ax.set_ylim(32, 34)
ax.set_xlabel('Pressure')
ax.set_ylabel('Salinity')
plt.legend()
plt.grid()
plt.savefig(os.path.join(drive,project,'salinity.png'),dpi=150)
```

```python
fig,ax = plt.subplots(figsize=[4,4])
pres = df["pressure"]
sal = df["salinity"]
temp = df["temperature"]
sc=ax.scatter(sal,temp,s=20,c=pres,cmap=cmo.haline_r,alpha=0.8,vmin=0, vmax=35)
plt.colorbar(sc,label='pressure')
# ax.set_ylim(5, 35)
# ax.set_ylim(32, 34)
ax.set_xlabel('Salinity')
ax.set_ylabel('Temperature')
plt.legend()
plt.grid()
plt.savefig(os.path.join(drive,project,'salinity.png'),dpi=150)
```
